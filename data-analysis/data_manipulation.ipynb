```python
#merge and join dataframes
# Customers DataFrame
df_customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Miami']
})

# Orders DataFrame
df_orders = pd.DataFrame({
    'OrderID': [101, 102, 103, 104],
    'CustomerID': [1, 3, 4, 5],
    'Product': ['Laptop', 'Smartphone', 'Desk Chair', 'Monitor'],
    'Quantity': [1, 2, 1, 1]
})
```

```python
df_customers
```

```python
df_orders
```

```python
# Inner merge (only matching CustomerID in both)
inner_merge = pd.merge(df_customers, df_orders, on='CustomerID', how='inner')

inner_merge
```

```python
# Left merge (all customers, orders matched where possible)
left_merge = pd.merge(df_customers, df_orders, on='CustomerID', how='left')
left_merge
```

```python
# Outer merge (all records from both, NaN where no match)
outer_merge = pd.merge(df_customers, df_orders, on='CustomerID', how='outer')
outer_merge
```
