```python
#data mamipulation

def assign_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Average'].apply(assign_grade)
```

```python
df
```

```python
#remove a column
print(df)
print(df.drop('Grade', axis=1)) #not persistant
print(df) #again appears
print(df.drop('Grade', axis=1, inplace=True)) # persistant
print(df)
```
